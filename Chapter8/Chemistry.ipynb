```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,21)
plt.figure(figsize=(10,8))
Boiling_1 = np.random.uniform(0.1,0.7,20)
Boiling_2 = np.random.uniform(0.1,0.7,20)

plt.bar(x, +Boiling_1, facecolor='#C0CA33', 
        edgecolor='white',label = "Boiling Point-1") #specify edgecolor by name
plt.bar(x, -Boiling_2, facecolor='#FF9800', 
        edgecolor='white',label = "Boiling Point-2")
for x,y in zip(x,Boiling_1):
  plt.text(x, y+0.05, '%.2f' % y, ha='center' , va= 'bottom', fontsize = 10)

plt.xlim(0,21)
plt.ylim(-1.25,+1.25)
plt.title ('Boiling Point Selection of Inorganic compounds')
plt.legend()
plt.show()
```

```python
data=Enegry
data = data.drop(['Materials Id'], axis=1)
data.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

numerical_transformer = StandardScaler()
label_transformer = OrdinalEncoder()

n_cols = [c for c in data.columns if data[c].dtype in ['int64', 'float64', 'int32', 'float32']]
obj_cols = [c for c in data.columns if data[c].dtype in ['object', 'bool']]
print(n_cols, obj_cols)

ct = ColumnTransformer([('num', numerical_transformer, n_cols), ('non_num', label_transformer, obj_cols),])
processed = ct.fit_transform(data)
new_data = pd.DataFrame(columns=data.columns, data=processed)
new_data.head()
```
